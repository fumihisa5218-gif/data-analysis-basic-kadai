```
import numpy as np
import matplotlib.pyplot as plt
```

```
#グラフ枠の設定
fig, axes = plt.subplots(2, 3,tight_layout = True)

#折れ線グラフの作成
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

# sin関数グラフの描画
x = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x)

axes[0,1].plot(x, y1, linewidth=2.0)

axes[0,1].set(
    xlim=(0, 10),
    xticks=np.arange(0, 10),
    ylim=(-1, 5),
    yticks=np.arange(-1, 6)
)

# ヒストグラムの作成
x = np.random.normal(15, 5, 2000)

# [0,2] にヒストグラムを描く
axes[0,2].hist(x, bins=30)

# x軸の範囲設定
axes[0,2].set_xlim(-10, 40)

#散布図の作成
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 60)
y = 4 + np.random.normal(0, 2, len(x))

# プロットする
axes[1,0].scatter(x, y)

plt.show()

```
