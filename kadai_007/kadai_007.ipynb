```python
import japanize_matplotlib
import matplotlib.pyplot as plt
```

```python
def category_count_and_plot(df):
    counts = df["カテゴリー"].value_counts()

    fig, ax = plt.subplots()
    ax.bar(counts.index.astype(str), counts.values)

    ax.set_xlabel("カテゴリー")
    ax.set_ylabel("出現頻度")
    ax.set_title("カテゴリー別出現頻度")

    plt.tight_layout()
    plt.show()
```

```python
category_count_and_plot(df)
```
